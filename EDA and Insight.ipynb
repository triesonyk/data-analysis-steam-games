## import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## load dataset

```python
df = pd.read_csv('for_EDA.csv')
df.sample(10)
```

```python
df.info()
```

We have dataset that contain all game from Steam, we will make a couple of insights which will be done on 4 different dataset that is taken from the main dataset. Those dataset will be:
- All game dataset
- High rated game dataset
- Games developed by giant publisher/developer
- Games developed by Indonesian developer
- Games developer by Indie developer

## EDA and Insight from All Games on Steam

### Descriptive Statistics

Observations :
- The average rating of a game is 76.4, the median is 80
- The average number of language supported by the game is 4+, the median is 2
- The average price of video game on PC is around 83k rupiah while the median is around 53k rupiah
- Most games has not been reviewed by more than 10 people, hence no rating has been given yet
- Most games need 4 GB of RAM to run
- Most games was released on the month of March, which is weird, holiday seasons like christmas or summer break should be the best time to release new game

```python
df.describe()
```

```python
df.describe(include='object')
```

### Univariate Analysis

#### Distributions

Observations :
- Most observations has been called on the previous Descriptive Analysis
- Contrary to what people believe, the price of a video game is actually quite cheap. Only AAA games or game that is published or developed by big company like Ubisoft, EA, Bethesda, etc are expensive (We will try to prove this on the next dataset). We can see this from the kdeplot and also the previous descriptive analysis (mean and median).
- From `total_reviews` we can see that a lot of people doesn't bother to put review, it should be different for games that is developed by giant developer/publisher
- Video games only has been around for a couple of decades, but we can see steady increase of number of games released every year

```python
num = ['year','user_rating','total_review','supported_language','price','pegi_rated','windows','mac','linux','VR']
cat = ['overall_reviews','ram','month']
```

```python
plt.figure(figsize=(12,10))
for i in range(0, len(num)):
    plt.subplot(4, 3, i+1)
    sns.kdeplot(x=df[num[i]], color='green')
    plt.xlabel(num[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(15,5))
sns.kdeplot(x=df['price'])
```

```python
plt.figure(figsize=(15,5))
sns.kdeplot(x=df['total_review'])
```

#### Box Plot

Observations :
- There are not many games on Steam that are released before 2010
- There are some games that have a really low rating on steam, probably because of bugs, glitch or other technical issues, or maybe sometimes the game is overpriced
- Only few games that has been reviewed by more than 1 million people
- Some games supported more than 15 languages
- Some games is priced so high, it is possible that this is a bundle consist of more than 1 game
- Funnily enough some game doesn't support Windows

```python
plt.figure(figsize=(10,10))
for i in range(0, len(num)):
    plt.subplot(2, 5, i+1)
    sns.boxplot(y=df[num[i]], color='green', orient='v')
    plt.tight_layout()
```

#### Count Plot

Observations :
- Most games that have already been reviewed more than 10 people have Positive, Mostly Positive, Very Positive and Mixed Reviews. It means that most games actually met the standard of most gamers
- Most games need at least 2 GB or 4 GB of RAM to run
- From count plot we can see that there is no significant difference of number of games released every month

```python
plt.figure(figsize=(20,8))
for i in range(0, len(cat)):
    plt.subplot(3, 1, i+1)
    sns.countplot(x = df[cat[i]], color='green', orient='v')
    plt.tight_layout()
```

### Features with List value

Observations :
- Giant Publisher/Developer like Ubisoft, KOEI, CAPCOM, Square Enix have been involved in a lot of video games
- Lot of games on steam were made by indie developers, in fact it is the number one tags that appear the most 
- Action, adventure and casual are the most popular genre on steam
- There are more single player game than multiplayer game

```python
cat_list = ['developer','publisher','tags']
for i in cat_list:
    df[i] = df[i].apply(eval)
```

```python
def to_1D(series): #
 return pd.Series([x for _list in series for x in _list])
```

```python
to_1D(df['developer']).value_counts().head(30)
```

```python
to_1D(df['publisher']).value_counts().head(30)
```

```python
to_1D(df['tags']).value_counts().head(30)
```

### Multivariate Analysis

#### Correlation

Observations :
- There is almost no correlation between each of numerical features
- Not a lot of information can be derived from pairplot
- Categorical features doesn't really have any effects on `price`, `total_review` and `user_rating`
- From kdeplot of `year` based on `ram` we can see that there is a different shape of distribution, it means that the latest game require more ram or higher spec compared to older game

```python
df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), cmap='Blues', annot=True, fmt='.2f')
```

```python
plt.figure(figsize=(15, 15))
sns.pairplot(df, diag_kind='kde')
```

```python
plt.figure(figsize=(20,8))
for i in range(0, len(cat)):
    plt.subplot(3, 1, i+1) 
    sns.kdeplot(data=df, x='user_rating', hue=cat[i])
    #plt.xlabel(features[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(20,8))
for i in range(0, len(cat)):
    plt.subplot(3, 1, i+1) 
    sns.kdeplot(data=df, x='price', hue=cat[i])
    #plt.xlabel(features[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(20,8))
for i in range(0, len(cat)):
    plt.subplot(3, 1, i+1) 
    sns.kdeplot(data=df, x='total_review', hue=cat[i])
    #plt.xlabel(features[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(20,8))
for i in range(0, len(cat)):
    plt.subplot(3, 1, i+1) 
    sns.kdeplot(data=df, x='year', hue=cat[i])
    #plt.xlabel(features[i])
    plt.tight_layout()
```
